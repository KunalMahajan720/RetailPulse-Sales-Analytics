```python
import pandas as pd
import numpy as np
```

```python

df = pd.read_csv('../cleaned_data/retail_clean.csv', 
                 parse_dates=['InvoiceDate'])

print(f"Data loaded: {len(df):,} rows")
print(f"Date range: {df['InvoiceDate'].min().date()} → {df['InvoiceDate'].max().date()}")
```

```python
# Revenue metrics 
total_revenue     = df['TotalRevenue'].sum()
total_orders      = df['InvoiceNo'].nunique()
total_customers   = df['CustomerID'].nunique()
total_products    = df['StockCode'].nunique()
avg_order_value   = total_revenue / total_orders
```

```python
# Peak month
monthly = df.groupby('YearMonth')['TotalRevenue'].sum()
peak_month        = monthly.idxmax()
peak_revenue      = monthly.max()
lowest_month      = monthly.idxmin()
lowest_revenue    = monthly.min()
```

```python
# Top country
country_rev       = df.groupby('Country')['TotalRevenue'].sum()
top_country       = country_rev.idxmax()
top_country_pct   = country_rev.max() / total_revenue * 100
second_country    = country_rev.nlargest(2).index[1]
```

```python
# Top product 
product_rev       = df.groupby('Description')['TotalRevenue'].sum()
top_product       = product_rev.idxmax()
top_product_rev   = product_rev.max()
```

```python
# Customer metrics 
order_freq        = df.groupby('CustomerID')['InvoiceNo'].nunique()
repeat_customers  = (order_freq > 1).sum()
onetime_customers = (order_freq == 1).sum()
repeat_pct        = repeat_customers / total_customers * 100
top_customer_id   = df.groupby('CustomerID')['TotalRevenue'].sum().idxmax()
top_customer_rev  = df.groupby('CustomerID')['TotalRevenue'].sum().max()
```

```python
# Time patterns 
peak_day          = df.groupby('DayOfWeek')['TotalRevenue'].sum().idxmax()
peak_hour         = df.groupby('Hour')['TotalRevenue'].sum().idxmax()
low_day           = df.groupby('DayOfWeek')['TotalRevenue'].sum().idxmin()

print("All metrics calculated successfully!")
```
